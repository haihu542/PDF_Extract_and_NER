```python
import fitz # PyMuPDF
import spacy
from spacy import displacy
import warnings
warnings.filterwarnings('ignore')
```

```python
# Step 1: Extract text from a PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    doc.close()
    return text
```

```python
# Step 2: Perform Named Entity Recognition using spaCy
def perform_ner(text):
    nlp = spacy.load("en_core_web_sm")
    return nlp(text)
```

```python
# Step 3: Visualize using displacy
def visualize_entities(doc):
    displacy.serve(doc, style="ent")
```

```python
# Example usage
if __name__ == "__main__":
    pdf_path = "We-Didnt-Start-the-Fire-lyrics.pdf" # Replace with your PDF file
    extracted_text = extract_text_from_pdf(pdf_path)
    doc = perform_ner(extracted_text)
    visualize_entities(doc)
```
